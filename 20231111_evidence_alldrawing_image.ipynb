```python
# 画像を回転させる関数
def rotate_img(img, rotate_type):
    if rotate_type == 0:
        img_rotate = img
    elif rotate_type == 90:
        img_rotate = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
    elif rotate_type == 180:
        img_rotate = cv2.rotate(img, cv2.ROTATE_180)
    elif rotate_type == 270:
        img_rotate = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
            
    return img_rotate
```

```python
# ドローイングプロセスビューワでの画像を回転角分だけ回転させて保存する
df_all = pd.read_excel('../data/all_drawing_info.xlsx')
folder_name_viewer_rotate = '20231110_drawingimg_viewer_rotate'

for drawing_id, rotate_type in zip(df_all['drawing_id'], df_all['rotate_type']):
    file_name = f'../temp/{folder_name_viewer_rotate}/drawing_id_{drawing_id}.jpg'
    img = cv2.imread(f'../temp/20231110_drawingimg_viewer_notrotate/drawing_id_{drawing_id}.jpg')
    img_rotate = rotate_img(img, rotate_type)
    cv2.imwrite(file_name, img_rotate)
    print(file_name)
```

```python
# 画像サイズを小さくして保存する
df_asai_target = pd.read_excel('../data/asai_target_info.xlsx', index_col=0)


for drawing_id, rotate_type in zip(df_asai_target['drawing_id'], df_asai_target['rotate_type']):
    save_file_name = f'../temp/20231110_drawingimg_plot_rotate_resize/drawing_id_{drawing_id}.jpg'
    img = cv2.imread(f'../temp/20231110_drawingimg_plot_rotate/drawing_id_{drawing_id}.jpg')
    
    if rotate_type == 0 or rotate_type == 180:
        img_resize = cv2.resize(img, (90, 63))
    elif rotate_type == 90 or rotate_type == 270:
        img_resize = cv2.resize(img, (63, 90))
    
    cv2.imwrite(save_file_name, img_resize)
    print(save_file_name)
```

```python
df_asai_target = pd.read_excel('../data/asai_target_info.xlsx', index_col=0)
vtr = imgsim.Vectorizer()
img_similarity = []

for drawing_id in df_asai_target['drawing_id']:
    file_name_plot = f'../temp/20231110_drawingimg_plot_rotate/drawing_id_{drawing_id}.jpg'
    file_name_viewer = f'../temp/20231110_drawingimg_viewer_rotate/drawing_id_{drawing_id}.jpg'

    img_plot = cv2.imread(file_name_plot)
    img_viewer = cv2.imread(file_name_viewer)
    
    img_plot_vec = vtr.vectorize(img_plot)
    img_viewer_vec = vtr.vectorize(img_viewer)
    
    dist = imgsim.distance(img_plot_vec, img_viewer_vec)
    img_similarity.append(dist)
    print(drawing_id)

df_similarity = pd.DataFrame({'drawing_id': df_asai_target['drawing_id'],
                              'img_similarity': img_similarity})
```
