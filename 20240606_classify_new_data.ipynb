```python
drawing_info = pd.read_excel('../data/all_drawing_info.xlsx', index_col=0).query('is_asai == 1')
# 和井田さん
waita_info  = drawing_info.query('name == "和井田瞳"').query('motif == "紙箱" or motif == "紙袋"')
# 佐野さん
sano_info  = drawing_info.query('name == "佐野りりこ"').query('motif == "紙箱" or motif == "紙袋"')
# 伊藤さん
ito_info  = drawing_info.query('name == "伊藤允彦"').query('motif == "紙箱" or motif == "紙袋"')
# 永松さん
nagamatsu_info  = drawing_info.query('name == "永松若葉"').query('motif == "紙箱" or motif == "紙袋"')
```

##### ▶SVM

```python
### 特徴量関連パラメータ ###
SAMPLE_TYPE = 'smaller_cossim'
SAMPLE_SIZE = 16
FEATURE_NAME = 'startend_seg_cossim'

# それぞれのデータを生成
waita_data = []
sano_data = []
ito_data = []
nagamatsu_data = []

# データ生成
for i, info in waita_info.iterrows():
    drawing_id = info['drawing_id']
    tsv_path = f'../data/all_tsv/id_{drawing_id}.tsv'
    rotate_type = info['rotate_type']
    data = pf.get_pipelined_data(tsv_path, rotate_type, SAMPLE_TYPE, SAMPLE_SIZE)
    waita_data.append(data)
    print(i)

for i, info in sano_info.iterrows():
    drawing_id = info['drawing_id']
    tsv_path = f'../data/all_tsv/id_{drawing_id}.tsv'
    rotate_type = info['rotate_type']
    data = pf.get_pipelined_data(tsv_path, rotate_type, SAMPLE_TYPE, SAMPLE_SIZE)
    sano_data.append(data)
    print(i)

for i, info in ito_info.iterrows():
    drawing_id = info['drawing_id']
    tsv_path = f'../data/all_tsv/id_{drawing_id}.tsv'
    rotate_type = info['rotate_type']
    data = pf.get_pipelined_data(tsv_path, rotate_type, SAMPLE_TYPE, SAMPLE_SIZE)
    ito_data.append(data)
    print(i)
    
for i, info in nagamatsu_info.iterrows():
    drawing_id = info['drawing_id']
    tsv_path = f'../data/all_tsv/id_{drawing_id}.tsv'
    rotate_type = info['rotate_type']
    data = pf.get_pipelined_data(tsv_path, rotate_type, SAMPLE_TYPE, SAMPLE_SIZE)
    nagamatsu_data.append(data)
    print(i)
```

##### ▶▶推論

```python
# 推論結果を格納する
waita_prediction = []
sano_prediction = []
ito_prediction = []
nagamatsu_prediction = []

# 和井田さん
for data in waita_data:
    # モデルの読み込み
    with open('../temp/model_and_scaler/svm_best_score/model.pkl', mode='rb') as f:
        model = pickle.load(f)
    # 推論
    X = pf.get_only_pipelined_feature(data, FEATURE_NAME)
    norm_X = pf.scale_minmax(X, SAMPLE_TYPE, SAMPLE_SIZE, FEATURE_NAME)
    np_norm_X = norm_X.to_numpy()
    prediction = model.predict(X)
    waita_prediction.append(prediction)
    
# 佐野さん
for data in sano_data:
    # モデルの読み込み
    with open('../temp/model_and_scaler/svm_best_score/model.pkl', mode='rb') as f:
        model = pickle.load(f)
    # 推論
    X = pf.get_only_pipelined_feature(data, FEATURE_NAME)
    norm_X = pf.scale_minmax(X, SAMPLE_TYPE, SAMPLE_SIZE, FEATURE_NAME)
    np_norm_X = norm_X.to_numpy()
    prediction = model.predict(X)
    sano_prediction.append(prediction)
    
# 伊藤さん
for data in ito_data:
    # モデルの読み込み
    with open('../temp/model_and_scaler/svm_best_score/model.pkl', mode='rb') as f:
        model = pickle.load(f)
    # 推論
    X = pf.get_only_pipelined_feature(data, FEATURE_NAME)
    norm_X = pf.scale_minmax(X, SAMPLE_TYPE, SAMPLE_SIZE, FEATURE_NAME)
    np_norm_X = norm_X.to_numpy()
    prediction = model.predict(X)
    ito_prediction.append(prediction)
    
# 永松さん
for data in nagamatsu_data:
    # モデルの読み込み
    with open('../temp/model_and_scaler/svm_best_score/model.pkl', mode='rb') as f:
        model = pickle.load(f)
    # 推論
    X = pf.get_only_pipelined_feature(data, FEATURE_NAME)
    norm_X = pf.scale_minmax(X, SAMPLE_TYPE, SAMPLE_SIZE, FEATURE_NAME)
    np_norm_X = norm_X.to_numpy()
    prediction = model.predict(X)
    nagamatsu_prediction.append(prediction)
```

##### ▶元のdfと結果の統合

```python
waita_data_pred = []
sano_data_pred = []
ito_data_pred = []
nagamatsu_data_pred = []

for data, pred in zip(waita_data, waita_prediction):
    is_good_svm = pd.DataFrame({'is_good_svm': pred})
    waita_data_pred.append(pd.concat([data, is_good_svm], axis=1))

for data, pred in zip(sano_data, sano_prediction):
    is_good_svm = pd.DataFrame({'is_good_svm': pred})
    sano_data_pred.append(pd.concat([data, is_good_svm], axis=1))

for data, pred in zip(ito_data, ito_prediction):
    is_good_svm = pd.DataFrame({'is_good_svm': pred})
    ito_data_pred.append(pd.concat([data, is_good_svm], axis=1))
    
for data, pred in zip(nagamatsu_data, nagamatsu_prediction):
    is_good_svm = pd.DataFrame({'is_good_svm': pred})
    nagamatsu_data_pred.append(pd.concat([data, is_good_svm], axis=1))
    
print(len(waita_data_pred))
print(len(sano_data_pred))
print(len(ito_data_pred))
print(len(nagamatsu_data_pred))
```

##### ▶実際のストロークを描画する

```python
# URLを指定したドローイングの、指定した順番のストロークの形状を描画する関数
def draw_stroke(x, y, title, save_path):
    # # --- グラフ描画 ---#
    CANVAS_COLOR = 'black'
    STROKE_COLOR = 'white'
    
    STROKE_WIDTH = 3
    SCALE = 0.02
    STROKE_SCALE = 1
    X_MIN = 0
    X_MAX = 420
    Y_MIN = 0
    Y_MAX = 420
    
    plt.rcParams['font.family'] = 'MS Gothic'
    fig, ax = plt.subplots(figsize=((X_MAX - X_MIN)*SCALE, (Y_MAX - Y_MIN)*SCALE), facecolor='white')
    ax.set_facecolor(CANVAS_COLOR)
    
    # x = data.query('stroke_id == @stroke_id')['normalized_x'].iloc[-1]
    # y = data.query('stroke_id == @stroke_id')['normalized_y'].iloc[-1]
    # shape = data.query('stroke_id == @stroke_id')['shape_int'].iloc[-1]
    
    # 軸の大きさ
    MARGIN_MM = 5
    ax.set_xlim(X_MIN-MARGIN_MM, X_MAX+MARGIN_MM)
    ax.set_ylim(Y_MIN-MARGIN_MM, Y_MAX+MARGIN_MM)
    
    ax.plot(x, y, color=STROKE_COLOR, linewidth=STROKE_WIDTH)
    # ax.scatter(x_plot[1:], y_plot[1:], color=STROKE_COLOR, marker='o')
    ax.scatter(x[:1], y[:1], color='orange', s=200)
    
    ax.set_title(title)
    ax.invert_yaxis()
    plt.tick_params(labelbottom=False, labelleft=False, labelright=False, labeltop=False, 
                    bottom=False, left=False, right=False, top=False)
    plt.savefig(save_path)
    # plt.show()
    plt.close()
    return
```
