### 気分転換で、ストロークを2次元平面からx, y座標に分割してみる

```python
from preprocess_data import PreprocessData
import matplotlib.pyplot as plt
```

```python
# 井原さん 紙箱 2回目 加工なし
data_180 = PreprocessData('http://skill.shinshu-u.ac.jp/2013/4-18/20130417140951_18-4-0-0-0_strokes.txt/strokes/pointlists',
                        180)
```

```python
def plot_timecoord(df, stroke_cnt, is_x):
    df_plot = df[stroke_cnt:stroke_cnt+1]  
    SCALE = 0.01
    plt.rcParams['font.family'] = 'Times New Roman'
    X_COLOR = 'blue'
    Y_COLOR = 'green'
    FACE_COLOR = 'white'
    STROKE_WIDTH = 1
    FIG_SIZE = (420*SCALE, 297*SCALE)
    
    fig, ax = plt.subplots(facecolor=FACE_COLOR, figsize=(FIG_SIZE[0], FIG_SIZE[1]))
    ax.set_facecolor(FACE_COLOR)
    if is_x:
        plot_list = df_plot['ptx_mm'].to_list()[0]
    else:
        plot_list = df_plot['pty_mm'].to_list()[0]
    
    # xの時系列
    ax.set_xlim([0, len(plot_list)])
    ax.set_ylim([min(plot_list), max(plot_list)])
    if is_x:
        ax.plot(range(len(plot_list)), plot_list, color=X_COLOR, linewidth=STROKE_WIDTH, marker='.')
    else:
        ax.plot(range(len(plot_list)), plot_list, color=Y_COLOR, linewidth=STROKE_WIDTH, marker='.')
        ax.invert_yaxis()
        
        
    # ax.invert_yaxis()
    plt.show()
```

```python
stroke_cnt = 222
plot_timecoord(data_180.df_lerp, stroke_cnt, True)
plot_timecoord(data_180.df_lerp, stroke_cnt, False)
data_180.draw_binstroke(stroke_cnt)

# print(data_180.df_drop_point[0:1]['max_points'])
```
